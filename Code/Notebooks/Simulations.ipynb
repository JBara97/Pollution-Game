```python
def truncate_colormap(cmap, minval=0.0, maxval=1.0, n=100):
    new_cmap = colors.LinearSegmentedColormap.from_list(
        'trunc({n},{a:.2f},{b:.2f})'.format(n=cmap.name, a=minval, b=maxval),
        cmap(np.linspace(minval, maxval, n)))
    return new_cmap
```

# Fixed Strategy Experiments

```python
folder = 'Fixed Strat Experiments/Experiment 6/'
N = 50
R = 5
L = 50
M = [3,4,6]
D = [10,20,30,40]
Phi = [0.5,1,2,3,4,5]
T=200
Runs = 20
Phat = np.zeros([Runs,len(M),len(D),len(Phi),T+1])
for r in trange(Runs):
    for i,m in enumerate(M):
        for j,d in enumerate(D):
                for k,phi in enumerate(Phi):
                    world = pollution.World(L=L,N=N,D=d,phi=phi,R=5,M=m)
                    Phat[r,i,j,k,0] = world.per_capita()
                    for t in range(1,T+1):
                        world.step()
                        Phat[r,i,j,k,t] = world.per_capita()

np.savetxt(folder+'Phat.csv',Phat.reshape((Runs*len(M)*len(D)*len(Phi),T+1)))
```

```python
cmaps = ['Blues','Greens','Oranges','Reds']
alpha = [0.3,0.5,0.7]
for k,phi in enumerate(Phi):
    fig, (a0) = plt.subplots(figsize=(10,8))#,ncols= 4, gridspec_kw={'width_ratios': [10, 0.2,0.2,0.2]})
    for j,d in enumerate(D):
        colormap = plt.get_cmap(cmaps[j])
        a0.set_prop_cycle(color=[colormap(k) for k in alpha])
        for i,m in enumerate(M):
            
            a0.set_xlabel("Time (t)")
            a0.set_ylabel("$\hat{P}(t)-\hat{P}(0)$")
            p = a0.plot(range(T+1),
                        Phat[:,i,j,k,:].mean(axis=0)-Phat[:,i,j,k,0].mean(axis=0),
                        label=f'({m},{d},{phi})')
            
#             for r in range(Runs):
#                 plt.plot(range(T+1),Phat[r,i,j,k,:]-Phat[r,i,j,k,0],c=p[0].get_color(),lw=0.1)
#             plt.legend()
#     a0.set_xscale('log')
    divider = make_axes_locatable(a0)
    for count,a in enumerate(alpha):
        CMAPS = [plt.get_cmap(c) for c in cmaps]
        cmap = mpl.colors.ListedColormap([c(a) for c in CMAPS])
        bounds = [5,15,25,35,45]
        norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
        sm=plt.cm.ScalarMappable(cmap=cmap,norm=norm)
        if count==0:
            cax = divider.append_axes("right", size="6%", pad=0.1)
        else:
            cax = divider.append_axes("right", size="6%", pad=0)
        cb = plt.colorbar(sm,boundaries=bounds,ticks=[],cax=cax,drawedges=False)#,ticks=[10,20,30,40])
        cb.ax.set_xlabel(f"{M[count]}")
    
    cb.set_ticks([10,20,30,40])
    cb.set_label("Defectors (D)")
    cb.ax.text(-80,1.8,'Migration (M)')
    fig.savefig(folder+f'Delta Phat: N={N},Phi={phi}.png')
    plt.close()
```

# Counting Clusters and Size

```python
folder = 'CC Experiments/Experiment 12/'
if not os.path.isdir(folder):
    os.mkdir(folder)
    
N = 50
R = 5
L = 50
M = [7,8,9]
D = [10,20,30,40]
Phi = [5]
T=200
Runs = 50
CC = np.zeros([Runs,len(M),len(D),len(Phi),2]) # the final axis of size 2 is [number of clusters, mean size]
Phat = np.zeros([Runs,len(M),len(D),len(Phi),T+1])
for r in trange(Runs):
    for i,m in enumerate(M):
        for j,d in enumerate(D):
                for k,phi in enumerate(Phi):
                    world = pollution.World(L=L,N=N,D=d,phi=phi,R=5,M=m)
                    Phat[r,i,j,k,0] = world.per_capita()
                    for t in range(1,T+1):
                        world.step()
                        Phat[r,i,j,k,t] = world.per_capita()
                    ccs = world.observe_clusters()
                    if len(ccs)>0:
                        CC[r,i,j,k,:] = [len(ccs),sum([len(c) for c in ccs])/len(ccs)]
                    else:
                        CC[r,i,j,k,:] = [0,0]

np.savetxt(folder+'CC.csv',CC.reshape((Runs*len(M)*len(D)*len(Phi),2)))
np.savetxt(folder+'Phat.csv',Phat.reshape((Runs*len(M)*len(D)*len(Phi),T+1)))
```
